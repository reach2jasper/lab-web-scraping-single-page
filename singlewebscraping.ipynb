```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select(" p")
```

```python
soup.select("p")
```

```python
soup.select(".title")
```

```python
soup.select(".artist")
```

```python


num_iter=len(soup.select(".title"))
num_iter
```

```python
songlist=soup.select(".title")
artlist=soup.select(".artist")
```

```python
song=[]
artist=[]
for i in range(num_iter):
    song.append(songlist[i].get_text())
    artist.append(artlist[i].get_text())
    
print(song)
print(artist)
```

```python
data=pd.DataFrame({"song":song,"artist":artist})
```
