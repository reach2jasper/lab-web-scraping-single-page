## Working-With-APIs

### 1. BC Ferries API

```python
from pandas.io.json import json_normalize
import requests
```

```python
response = requests.get('https://bcferriesapi.ca/api/')
results = response.json()
```

```python
response.status_code
#API Fetching is successful
```

```python
results
```

```python
flattend_data=json_normalize(results)
flattend_data.T
```
